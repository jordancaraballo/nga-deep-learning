# Visualize Probability Array

Simple notebook to visualize the probability array per class. Nothing complicated here, just for testing purposes when dealing with semi-supervised learning.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
filename = '/Users/jacaraba/Desktop/CURRENT_PROJECTS/GMU/100_unet_mle5_128_431.h5/segments/number140_prob.npy'
NCLASSES = 5
probabilities = np.load(filename)
```

```python
probabilities
```

```python
fig = plt.figure(figsize=(15, 15))
grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(2, 3),  # creates 2x2 grid of axes
                 axes_pad=0.1,  # pad between axes in inch.
                 )

for ax, im_id in zip(grid, range(NCLASSES)):
    # Iterating over the grid returns the Axes.
    ax.imshow(probabilities[:,:,im_id])

plt.show()
```
